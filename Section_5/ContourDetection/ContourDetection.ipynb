```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
img = cv2.imread(r"C:\Users\umut3\Desktop\Computer-Vision-with-Python\DATA\internal_external.png")
img.shape
gray_img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
```

```python
plt.imshow(img, cmap='gray')
```

```python
contours, hierarchy = cv2.findContours(gray_img, cv2.RETR_CCOMP, cv2.CHAIN_APPROX_SIMPLE)
```

```python
external_c = np.zeros(img.shape)
external_c.shape
```

```python
for i in range(len(contours)):
    # External
    if hierarchy[0][i][3] == -1:

        cv2.drawContours(external_c, contours, i, 255, -1)
```

```python
plt.imshow(external_c.astype("uint8"), cmap='gray')

internal_c = np.zeros(img.shape)
```

```python
for i in range(len(contours)):
    # Internal
    if hierarchy[0][i][3] != -1:

        cv2.drawContours(internal_c, contours, i, 255, -1)
```

```python
plt.imshow(internal_c.astype("uint8"), cmap='gray')
```
