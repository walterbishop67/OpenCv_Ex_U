```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
face = cv2.imread(r"C:\Users\umut3\Desktop\Computer-Vision-with-Python\DATA\sammy_face.jpg")
plt.imshow(face)
face_gray = cv2.cvtColor(face, cv2.COLOR_BGR2GRAY)
```

```python
edges = cv2.Canny(image=face_gray, threshold1=127, threshold2=127)
plt.imshow(edges)
```

```python
med_val = np.median(face)
lower = int(max(0, 0.7*med_val)) # % 70
upper = int(max(255, 1.3*med_val)) # %130
```

```python
edges = cv2.Canny(image=face_gray, threshold1=lower, threshold2=upper)
plt.imshow(edges)
```

```python
blurred_img = cv2.blur(face, ksize=(3,3))
```

```python
edges = cv2.Canny(image=blurred_img, threshold1=lower, threshold2=upper-30)
plt.imshow(edges)
```
