```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
flat_chess = cv2.imread(r"C:\Users\umut3\Desktop\Computer-Vision-with-Python\DATA\flat_chessboard.png")
flat_chess = cv2.cvtColor(flat_chess, cv2.COLOR_BGR2RGB)

real_chess = cv2.imread(r"C:\Users\umut3\Desktop\Computer-Vision-with-Python\DATA\real_chessboard.jpg")
real_chess = cv2.cvtColor(real_chess, cv2.COLOR_BGR2RGB)

flat_chess_gray = cv2.cvtColor(flat_chess, cv2.COLOR_BGR2GRAY)
real_chess_gray = cv2.cvtColor(real_chess, cv2.COLOR_BGR2GRAY)
```

```python
corners = cv2.goodFeaturesToTrack(flat_chess_gray, 64, 0.01, 10)
```

```python
corners = np.intp(corners)
```

```python
for i in corners:
    x,y = i.ravel()
    cv2.circle(flat_chess, (x,y), 3, (255, 0, 0), -1)
```

```python
plt.imshow(flat_chess)
```

```python
corners = cv2.goodFeaturesToTrack(real_chess_gray, 100, 0.01, 10)
```

```python
corners = np.intp(corners)
```

```python
for i in corners:
    x,y = i.ravel()
    cv2.circle(real_chess, (x,y), 3, (255, 0, 0), -1)
```

```python
plt.imshow(real_chess)
```
