```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
img = cv2.imread(r"C:\Users\umut3\PycharmProjects\pythonProject4\dog.4003.jpg")
plt.imshow(img)
```

```python
img_rbg = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img_rbg)
```

```python
new_img = img_rbg.copy()
new_img = cv2.flip(new_img, 0)
plt.imshow(new_img)
```

```python
cv2.rectangle(img=img_rbg,pt1=(65,20), pt2=(210,210), color=(0,0,255), thickness=5)
plt.imshow(img_rbg)
img = cv2.imread(r"C:\Users\umut3\PycharmProjects\pythonProject4\dog.4003.jpg")
```

```python
vertices = np.array([ [120, 250], [180, 250], [150, 200] ], np.int32)
vertices.shape

pts = vertices.reshape((-1, 1, 2))
vertices

cv2.polylines(img_rbg, [pts], isClosed=True, color=(0,0,255), thickness=10)
plt.imshow(img_rbg)

cv2.fillPoly(img_rbg, [pts], color=(0,0,255))
plt.imshow(img_rbg)
```

```python
def create_circle (event, x, y, flags, param):
    if event == cv2.EVENT_LBUTTONDOWN:
        cv2.circle(img, (x,y), 20, (255, 0, 0), thickness=10)

cv2.namedWindow(winname='dog')
cv2.setMouseCallback('dog', create_circle)

while True:
    cv2.imshow('dog', img)

    if cv2.waitKey(20) & 0xFF == 27:
        break

cv2.destroyAllWindows()
```
