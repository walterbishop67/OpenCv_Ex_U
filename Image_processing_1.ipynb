```python
import cv2
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
img1 = cv2.imread(r"C:\Users\umut3\PycharmProjects\CV\dog.4006.jpg")
img2 = cv2.imread(r"C:\Users\umut3\PycharmProjects\CV\cat.4001.jpg")
```

```python
img1 = cv2.cvtColor(img1, cv2.COLOR_BGR2RGB)
plt.imshow(img1)
img1.shape
```

```python
img2 = cv2.cvtColor(img2, cv2.COLOR_BGR2RGB)
plt.imshow(img2)
img2.shape
```

```python
imgg1 = cv2.resize(img1, (450, 500))
imgg2 = cv2.resize(img2, (450, 500))

blended = cv2.addWeighted(src1=imgg1, alpha=0.5, src2=imgg2, beta=0.5, gamma=0)
plt.imshow(blended)
```

```python
imggg1 = cv2.resize(img1, (135 ,250))
plt.imshow(imggg1)

x_offset = 0
y_offset = 0

x_end = x_offset + imggg1.shape[1]
y_end = y_offset + imggg1.shape[0]

img2[y_offset:y_end, x_offset:x_end] = imggg1
plt.imshow(img2)
```
