```python
img1 = cv2.imread(r"C:\Users\umut3\PycharmProjects\CV\dog_backpack.png")
img2 = cv2.imread(r"C:\Users\umut3\PycharmProjects\CV\watermark_no_copy.png")

img1 = cv2.cvtColor(img1, cv2.COLOR_BGR2RGB)
plt.imshow(img1)
img1.shape

img2 = cv2.cvtColor(img2, cv2.COLOR_BGR2RGB)
plt.imshow(img2)
img2.shape
```

```python
x_offset = 934 - 600
y_offset = 1401 - 600

rows, columns, channels = img2.shape
```

```python
roi = img1[y_offset:1401, x_offset:934]
plt.imshow(roi)
```

```python

img2 = cv2.resize(img2, (600, 600))
plt.imshow(img2)
```

```python
img2gray = cv2.cvtColor(img2, cv2.COLOR_BGR2GRAY)
plt.imshow(img2gray)
```

```python
mask_inv = cv2.bitwise_not(img2gray)
plt.imshow(mask_inv, cmap='gray')
white_background = np.full(img2.shape, 255, dtype=np.uint8)
bk = cv2.bitwise_or(white_background, white_background, mask=mask_inv)
plt.imshow(bk)
```

```python
fg = cv2.bitwise_or(img2, img2, mask=mask_inv)
plt.imshow(fg)
```

```python
final_roi = cv2.bitwise_or(roi, fg)
plt.imshow(final_roi)
```

```python
large_img = img1
small_img = final_roi
```

```python
large_img[y_offset:y_offset+small_img.shape[0],x_offset:x_offset+small_img.shape[1]] = small_img
plt.imshow(large_img)
```
