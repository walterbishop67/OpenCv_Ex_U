```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
gorilla = cv2.imread(r"C:\Users\umut3\Desktop\Computer-Vision-with-Python\Computer-Vision-with-Python\DATA\gorilla.jpg",0)
```

```python
def display(img, cmap=None):
    fig = plt.figure(figsize=(10,8))
    ax = fig.add_subplot(111)
    ax.imshow(img, cmap)
```

```python
display(gorilla, cmap='gray')
```

```python
gorilla.shape
```

```python
hist_values = cv2.calcHist([gorilla], channels=[0], mask=None, histSize=[256], ranges=[0, 256])
```

```python
plt.plot(hist_values)
```

```python
eq_gorilla = cv2.equalizeHist(gorilla)
```

```python
display(eq_gorilla, 'gray')
```

```python
hist_values = cv2.calcHist([eq_gorilla], channels=[0], mask=None, histSize=[256], ranges=[0, 256])
```

```python
plt.plot(hist_values)
```
