```python
img2 = cv2.imread(r"C:\Users\umut3\Desktop\Computer-Vision-with-Python\Computer-Vision-with-Python\DATA\crossword.jpg", 0)
plt.imshow(img2, cmap='gray')
```

```python
def show_pic(img):
    fig = plt.figure(figsize=(15,15))
    ax = fig.add_subplot(111)
    ax.imshow(img, cmap='gray')

show_pic(img2)
```

```python
ret, th1 = cv2.threshold(img2, 175, 255, cv2.THRESH_BINARY)
show_pic(th1)
```

```python
th2 = cv2.adaptiveThreshold(img2, 255, cv2.ADAPTIVE_THRESH_MEAN_C, cv2.THRESH_BINARY, 11, 8)
show_pic(th2)
```

```python
blended = cv2.addWeighted(src1=th1, alpha=0.6, src2=th2, beta=.4, gamma=0)
show_pic(blended)
```
