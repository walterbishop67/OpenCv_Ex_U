```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
rainbow = cv2.imread(r'C:\Users\umut3\Desktop\Computer-Vision-with-Python\Computer-Vision-with-Python\DATA\rainbow.jpg')
show_img = cv2.cvtColor(rainbow, cv2.COLOR_BGR2RGB)
```

```python
img = rainbow
img.shape
```

```python
mask = np.zeros(img.shape[:2], np.uint8)
```

```python
plt.imshow(mask, cmap='gray')
```

```python
mask[300:400, 100:400] = 255
plt.imshow(mask, cmap='gray')
```

```python
plt.imshow(show_img)
```

```python
masked_img = cv2.bitwise_and(img, img, mask=mask)
```

```python
show_masked_img = cv2.bitwise_and(show_img, show_img, mask=mask)
```

```python
plt.imshow(show_masked_img)
```

```python
hist_mask_values_red = cv2.calcHist([rainbow], channels=[2], mask=mask, histSize=[256], ranges=[0,256])
```

```python
hist_values_red = cv2.calcHist([rainbow], channels=[2], mask=None, histSize=[256], ranges=[0,256])
```

```python
plt.plot(hist_mask_values_red)
plt.title('Red Histogram For Masked Rainbow')
```

```python
plt.plot(hist_values_red)
plt.title('Red Histogram For Normak Rainbow')
```
