```python
import cv2
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline

def display_img(img, cmap=None):
    fig = plt.figure(figsize=(12,10))
    ax = fig.add_subplot(111)
    ax.imshow(img, cmap)

img = cv2.imread(r"C:\Users\umut3\PycharmProjects\opennn\giraffes.jpg")
img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
display_img(img)

```

```python
img = cv2.imread(r"C:\Users\umut3\PycharmProjects\opennn\giraffes.jpg", 0)
ret, thresh1 = cv2.threshold(img, 127, 255, cv2.THRESH_BINARY)
```

```python
display_img(thresh1, cmap="gray")
```

```python
img = cv2.imread(r"C:\Users\umut3\PycharmProjects\opennn\giraffes.jpg")
img = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
display_img(img)
```

```python
img = cv2.imread(r"C:\Users\umut3\PycharmProjects\opennn\giraffes.jpg")
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
kernel = np.ones(shape=(4, 4),  dtype=np.float32)/10
result = cv2.filter2D(img, -1, kernel)

display_img(result)
```

```python
img = cv2.imread(r"C:\Users\umut3\PycharmProjects\opennn\giraffes.jpg", 0)
sobelx = cv2.Sobel(img, cv2.CV_64F, 1,0,ksize=5)
display_img(sobelx, cmap="gray")
```

```python
img = cv2.imread(r"C:\Users\umut3\PycharmProjects\opennn\giraffes.jpg")

color = ['b', 'g', 'r']

for i, col in enumerate(color):
    histr = cv2.calcHist([img], [i], None, [256], [0,256])
    plt.plot(histr, color=col)
```
