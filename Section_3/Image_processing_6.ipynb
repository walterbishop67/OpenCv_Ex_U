```python
import cv2
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
rainbow = cv2.imread(r"C:\Users\umut3\Desktop\Computer-Vision-with-Python\Computer-Vision-with-Python\DATA\rainbow.jpg")
show_rainbow = cv2.cvtColor(rainbow, cv2.COLOR_BGR2RGB)

plt.imshow(show_rainbow)

horse = cv2.imread(r"C:\Users\umut3\Desktop\Computer-Vision-with-Python\Computer-Vision-with-Python\DATA\horse.jpg")
show_horse = cv2.cvtColor(horse, cv2.COLOR_BGR2RGB)

plt.imshow(show_horse)

brick = cv2.imread(r"C:\Users\umut3\Desktop\Computer-Vision-with-Python\Computer-Vision-with-Python\DATA\bricks.jpg")
show_brick = cv2.cvtColor(brick, cv2.COLOR_BGR2RGB)

plt.imshow(show_brick)
```

```python
hist_values = cv2.calcHist([show_brick], channels=[1], mask=None, histSize=[256], ranges=[0, 256])
hist_values.shape

plt.plot(hist_values)
```

```python
img = show_brick

color = ('b', 'g', 'r')

for i, col in enumerate(color):
    histr = cv2.calcHist([img], [i], None, [256], [0,256])
    plt.plot(histr, color=col)
    plt.xlim([0,256])

plt.title('HISTOGRAM')
```

```python
img = show_horse

color = ('b', 'g', 'r')

for i, col in enumerate(color):
    histr = cv2.calcHist([img], [i], None, [256], [0,256])
    plt.plot(histr, color=col)
    plt.xlim([0,50])
    plt.ylim([0,500000])

plt.title('HISTOGRAM')
```
